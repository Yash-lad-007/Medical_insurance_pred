```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score
```

```python
import pandas as pd
df=pd.read_csv('/content/insurance.csv')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,10))
sns.histplot(df['bmi'],kde = True)
plt.title("BMI Distribution with KDE")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(10,4))
sns.boxplot(x = df['bmi'])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
# remove the outlier
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

print(f" lower :{lower_bound}")
print(f" upper :{upper_bound}")
```

```python
# remove duplicates
df.drop_duplicates(inplace = True)
```

```python
# Clipping the outliers
df_cleaned = df[(df['bmi'] >=lower_bound ) & (df['bmi'] <= upper_bound)]
```

```python
plt.figure(figsize=(15,10))
sns.histplot(df_cleaned['bmi'],kde = True)
plt.title("BMI Distribution with KDE")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(10,4))
sns.boxplot(x = df_cleaned['bmi'])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
df_cleaned
```

```python
# one hot encoding
df_cleaned = pd.get_dummies(df_cleaned,drop_first = True)
df_cleaned
```

```python
df_cleaned.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_cleaned.corr(),annot =True)
```

```python
# feature enginerring
df_cleaned.columns
```

```python
# x and y
x = df_cleaned[['age','bmi','smoker_yes']]
y = df_cleaned['charges']
```

```python
# train test splits
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state =23)
```

```python
#train the models
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)# testing on test data
y_train_pred = model.predict(x_train)# testing on train data
```

```python
print(f'testing results : {r2_score(y_test,y_pred)}')
print(f'training results : {r2_score(y_train,y_train_pred)}')
```

```python
# how to save a model
import pickle
with open("Best_model.pkl","wb") as f:
  pickle.dump(model,f)
```
